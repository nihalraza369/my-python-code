```python
count = 6
# Loop tab tak chalega jab tak count 5 ya usse chhota hai
while count >= 3:
    print(f"Count is: {count}")
    count += 1
      # Count mein 1 ka izafa kar rahe hain taake loop khatam ho sake
```

```python
try:
    number = int(input("Enter number: "))
    result = 10 / number
    print("Result is:",result)
except ZeroDivisionError:
    print("You cannot divide by zero!")
except ValueError:
    print("Please enter a valid number!")
finally:
    print("completed....")
```
